## 2nd Modified Files Clean

```python
import os
import pandas as pd
```

```python
second_folder = 'D:/COMSATS/FYP/Coding Part/Whole Process with Minor Changes/2nd Cleaned Files'
folder_list = os.listdir(second_folder)
folder_list
```

```python
dataframes = []

# Loop through the files in the folder
for filename in os.listdir(second_folder):
    if filename.endswith('.csv'):
        file_path = os.path.join(second_folder, filename)
        df = pd.read_csv(file_path)
        dataframes.append(df)
```

```python
combined_df = pd.concat(dataframes, ignore_index=True)
combined_df
```

```python
combined_df.isnull().sum()
```

```python
filtered_df = combined_df[pd.isna(combined_df['Car_Name'])]
filtered_df
```

```python
filtered_df['Car_Name'] = filtered_df['AD_Name'].str.split().str[1]
filtered_df.head()
```

```python
filtered_df['Brand_Name'] = filtered_df['AD_Name'].str.split().str[0]
filtered_df.head()
```

```python
combined_df = pd.concat((combined_df,filtered_df))
combined_df
```

```python
combined_df = combined_df.dropna(subset=['Car_Name'])
```

```python
combined_df['Car_Name'].unique()
```

```python
combined_df = combined_df[combined_df['Car_Name'] != 'Passo']
combined_df = combined_df[combined_df['Car_Name'] != 'Charade']
combined_df = combined_df[combined_df['Car_Name'] != 'Bluebird']
combined_df = combined_df[combined_df['Car_Name'] != 'March']
combined_df = combined_df[combined_df['Car_Name'] != 'Belta']
combined_df = combined_df[combined_df['Car_Name'] != 'Khyber']
combined_df = combined_df[combined_df['Car_Name'] != 'V2']
combined_df = combined_df[combined_df['Car_Name'] != 'Yaris']
```
