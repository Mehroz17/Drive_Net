### Now for Ad_Name ðŸ“‡
For ad name I am going to check where is the ad_name empty then go to that link and get the name of the ad and paste it

```python
def get_ad_name(ad_link):
    try:
        response = requests.get(ad_link)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            ad_name = soup.find('div',class_='well').find_next('h1').text
            print(ad_name)
            return ad_name
        else:
            return None
    except Exception as e:
        print("An error occurred:", e)
        return None

# Iterate through rows and fill empty AD_Name with scraped data
for index, row in combined_data.iterrows():
    if pd.isna(row['AD_Name']) or row['AD_Name'].strip() == '':
        ad_link = row['Ad_link']
        print(ad_link)
        ad_name = get_ad_name(ad_link)
        if ad_name:
            combined_data.at[index, 'AD_Name'] = ad_name
```
