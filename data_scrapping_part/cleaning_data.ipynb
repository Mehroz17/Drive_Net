<h2>Price column is done</h2>
Now its time for price

```python
dataset['Price']
```

```python
# replacing the un-wanted words , and converting the crore price in lacs
dataset = dataset[~dataset['Price'].str.contains("Call for price")] 
dataset["Price"] = dataset['Price'].str.replace('PKR','')
dataset["Price"] = dataset['Price'].str.replace("crore",'100')
dataset["Price"] = dataset['Price'].str.replace("lacs",'')
dataset.loc[:, 'Price'] = dataset['Price'].apply(lambda x: float(x.replace(' 100', '')) * 100 if '100' in str(x) else x)
```

```python
dataset["Price"] = dataset["Price"].astype('float')
```

```python
dataset.dtypes
```

Price is done now for Millage

```python
dataset["Milage"] = dataset["Milage"].str.replace(" km",'')
dataset["Milage"] = dataset["Milage"].str.replace(",",'')

dataset["Milage"] = dataset['Milage'].astype('int64')
```
